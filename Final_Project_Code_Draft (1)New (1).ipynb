## Trying to factor

```python
def EducationFunc(df1): 
    if (df1['Education']=='High School or GED' ): return 0
    elif (df1['Education']=='Some Phd' ): return 1
    elif (df1['Education']=='Completed Undergraduate' ): return 2
    elif (df1['Education']=='Some Undergraduate' ): return 3
    elif (df1['Education']=='Some\xa0Masters' ): return 4
    elif (df1['Education']=='Completed Masters' ): return 5
    elif (df1['Education']=='Completed Phd' ): return 6
    elif (df1['Education']=='Some highschool' ): return 7
    else: return 'Error'
df1['EducationNumerical'] = df1.apply(EducationFunc,axis=1) 
```

```python
def GenderFunc(df1): 
    if (df1['Gender']=='Female' ): return 0
    elif (df1['Gender']=='Male' ): return 1
    else: return 'Error'
df1['GenderNumerical'] = df1.apply(GenderFunc,axis=1) 
```

```python
def RegionFunc(df1): 
    if (df1['Region']=='Mountain' ): return 0
    elif (df1['Region']=='East South Central' ): return 1
    elif (df1['Region']=='Pacific' ): return 2
    elif (df1['Region']=='East North Central' ): return 3
    elif (df1['Region']=='South Atlantic' ): return 4
    elif (df1['Region']=='Middle Atlantic' ): return 5
    elif (df1['Region']=='New England' ): return 6
    elif (df1['Region']=='West South Central' ): return 7
    elif (df1['Region']=='West North Central' ): return 8
    else: return 'Error'
df1['RegionNumerical'] = df1.apply(RegionFunc,axis=1) 
```

```python
def DeviceTypeFunc(df1): 
    if (df1['Device Type']=='Android Phone / Tablet' ): return 0
    elif (df1['Device Type']=='MacOS Desktop / Laptop' ): return 1
    elif (df1['Device Type']=='iOS Phone / Tablet' ): return 2
    elif (df1['Device Type']=='Windows Desktop / Laptop' ): return 3
    elif (df1['Device Type']=='Other' ): return 4
    else: return 'Error'
df1['DeviceTypeNumerical'] = df1.apply(DeviceTypeFunc,axis=1) 
```

```python
def AgeFunc(df1): 
    if (df1['Age']=='30-44' ): return 0
    elif (df1['Age']=='18-29' ): return 1
    elif (df1['Age']=='45-60' ): return 2
    elif (df1['Age']=='> 60' ): return 3
    else: return 'Error'
df1['AgeNumerical'] = df1.apply(AgeFunc,axis=1) 
```

```python
df["Household Income"].unique()
```

```python
def HouseIncEdu(df1):
    if df1['Household Income'] == '$25,000-$49,999' : return 1 
    elif df1['Household Income'] == '$50,000-$74,999' : return 2 
    elif df1['Household Income'] == '$0-$9,999' : return 3 
    elif df1['Household Income'] == '$150,000-$174,999' : return 4
    elif df1['Household Income'] == '$100,000-$124,999' : return 5
    elif df1['Household Income'] == '$10,000-$24,999' : return 6
    elif df1['Household Income'] == '$125,000-$149,999' : return 7
    elif df1['Household Income'] == '$75,000-$99,999' : return 8
    elif df1['Household Income'] == '$200,000+' : return 9
    elif df1['Household Income'] == '$175,000-$199,999' : return 10
    elif df1['Household Income'] == 'Prefer not to answer' : return 11
    else: return 'Error'
df1['HouseIncEduNumerical'] = df1.apply(HouseIncEdu, axis = 1)
```
